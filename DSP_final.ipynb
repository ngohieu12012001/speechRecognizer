<h3 align='center'>CHƯƠNG 1 – TỔNG QUAN
</h3>

**1.1 Introduction**

Trong những năm gần đây các vấn đề về xử lý tiếng nói luôn là chủ đề được các nhà nghiên  cứu đặc biệt quan tâm. Đây cũng chính là tiền đề để xây dựng và phát triển các lĩnh vực liên  quan đến giọng nói như cao độ, âm sắc, màu sắc ngôn ngữ,...Từ đó hình thành các bài toán mang tính ứng dụng cao như tổng hợp giọng nói, nhận biết tiếng nói, điều chỉnh giọng nói,.. Vậy trong bài báo cáo này chúng ta sẽ trình bày về nhận biết tiếng nói.

*1.1.1 Khái niệm tiếng nói và nhận dạng tiếng nói*

Tiếng nói là một phương tiện trao  đổi thông tin tiện ích vốn có của con người. Tiếng nói xuất hiện dưới nhiều hình thức mà ta hay gọi là đàm thoại, tiếng nói giúp người nghe hiểu được ý nghĩa cần diễn đạt. Sự biểu đạt của tiếng nói đã hỗ trợ rất nhiều cho các hệ thống cần giao tiếp cần sử dụng đến tiếng nói. Nếu chúng ta phân tích quá trình giao tiếp qua nhiều lớp thì lớp thấp nhất chính là âm thanh và lớp cuối cùng nhằm để diễn đạt một ý nghĩa muốn nói.

![kntn](https://i.imgur.com/UVxwV3T.png)

Nhận dạng tiếng nói là một quá trình nhận dạng các mẫu, với mục đích là phân lớp  thông tin đầu vào là tín hiệu tiếng nói thành một dãy tuần tự các mẫu đã được học trước đó và lưu trữ trong bộtuk nhớ. Các mẫu là các đơn vị nhận dạng, chúng có thể là các từ, hoặc các âm vị. Việc nhận dạng sẽ diễn ra đơn giản nếu như các mẫu này bất biến và không thay đổi, nhưng các bài toán sử dụng các mẫu giống nhau như này sẽ không mang lại hiệu quả cao trong việc nhận dạng, vì tính chất tiếng nói của con người luôn biến thiên theo thời gian. Cũng chính tính chất cơ bản này mà bài toán nhận dạng tiếng nói sẽ trở nên khó khăn hơn. Một điều đặc biệt mà chúng ta cần lưu ý cho bài toán nhận dạng giọng nói rằng tốc độ nói, môi trường âm học,.. của mỗi người là khác nhau.

Các thành phần chính của hệ thống nhận dạng tiếng nói:
* Tiền xử lý, tín hiệu tiếng nói từ dạng liên tục sẽ được biến đổi thành dạng 
tín hiệu số, các tín hiệu số này sẽ được chia nhỏ thành các khung có kích thước 10ms đến 30ms.
* Trích xuất đặc trưng, Mỗi khung sẽ được trích xuất bằng các phương pháp như MFCC, LPC, STFT. Đầu ra là các vector đặc trưng tương ứng với các khung.
* Mô hình hóa ngữ âm (AM), quá trình huấn luyện mô hình này sẽ cho ta thấy sự liên kết giữa các đơn vị nhỏ nhất (âm vị) của tiếng nói với đặc trưng. Đây cũng là thành phần cơ bản của hệ thống nhận dạng tiếng nói để phản ánh sự tương quan giữa âm thanh và ngữ âm.
* Mô hình ngôn ngữ (LM) nhằm để tính toán xác suất của các cụm từ. Khi mô hình ngữ âm phát hiện ra nhiều từ có cùng xác suất, thì mô hình ngôn ngữ có nhiệm vụ chỉ ra từ gần nhất thông qua xác suất. 
* Mô hình nhận dạng là quá trình tham chiếu các mẫu chưa biết với các mẫu đã được tính toán và lưu trữ.

**1.2 Problem definition:**

*1.2.1 Nhận dạng giọng nói* còn gọi là speech recognition là quá trình biến đổi tín hiệu âm thanh thu được của người nói thành một chuỗi các từ có nội dung tương ứng dưới dạng văn bản, nhằm mở rộng thêm khả năng lưu trữ, tìm kiếm thông tin và khả năng tương tác với máy tính. Nhận dạng tiếng nói là quá trình phức tạp bao gồm nhiều khâu biến đổi. Tín hiệu mà người phát ra là tín hiệu tương tự, qua quá trình lấy mẫu, lượng tử hoá và mã hoá để thu được các mẫu tín hiệu dạng số (tín hiệu mà máy tính có thể hiểu và xử lý được).

![PD1](https://i.imgur.com/e5ZDarg.png)

Nhận diện giọng nói là một bài toán trong lĩnh vực xử lý tiếng nói, trong đó mục tiêu là phân loại âm thanh đầu vào thành các từ hoặc câu nói tương ứng với ngôn ngữ đã cho trước. Bài toán này có rất nhiều ứng dụng trong cuộc sống thực, chẳng hạn như hệ thống nhận diện giọng nói để mở khóa điện thoại, hệ thống điều khiển bằng giọng nói, hoặc hệ thống chuyển đổi giọng nói thành văn bản.


*1.2.2 Vấn đề xử lý tiếng nói* là quá trình phân tích, xử lý và tạo ra các tín hiệu âm thanh và giọng nói. Nó là một lĩnh vực quan trọng trong trí tuệ nhân tạo và khoa học máy tính, với nhiều ứng dụng trong các hệ thống truyền thông, tương tác giọng nói, giải trí và an ninh. Bài toán xử lý tiếng nói bao gồm:
- Nhận dạng giọng nói: là quá trình chuyển đổi giọng nói thành văn bản. Bài toán này rất quan trọng trong các ứng dụng như tương tác giọng nói và trợ lý ảo.
- Nhận diện người nói: là quá trình xác định danh tính của người nói từ tín hiệu giọng nói. Bài toán này có ứng dụng trong các hệ thống xác thực người dùng.

*1.2.2.1 Nhận diện người nói (Speaker Recognition)*: là một lĩnh vực của xử lý tiếng nói, có nhiệm vụ xác định danh tính của người nói trong một bài nói hoặc tín hiệu giọng nói. Bài toán này có rất nhiều ứng dụng trong thực tế như: xác thực người dùng trong hệ thống đăng nhập, giám sát giọng nói trong các cuộc gọi điện thoại, ghi âm tòa án,... 

*Dữ liệu đầu vào* của bài toán nhận diện tiếng nói là tín hiệu âm thanh thu được từ microphone hoặc từ các tệp âm thanh, có thể là các file WAV. Tuy nhiên, trước khi sử dụng cho bài toán nhận diện tiếng nói, tín hiệu âm thanh thường phải trải qua một quá trình tiền xử lý để chuẩn hóa dữ liệu:

- Tín hiệu giọng nói: Đây là dữ liệu đầu vào chính cho bài toán nhận diện người nói. Tín hiệu giọng nói có thể được thu thập bằng các thiết bị thu âm, chẳng hạn như microphone hoặc bộ đàm, hoặc từ các bản ghi âm trước đó. 
- Thông tin về người nói: Để cải thiện tính chính xác của hệ thống nhận diện người nói, các thông tin về người nói cũng có thể được sử dụng làm dữ liệu đầu vào, chẳng hạn như: Tên, giới tính, độ tuổi, ngôn ngữ, vùng miền, và những đặc điểm khác của giọng nói.

*Dữ liệu đầu ra*: đối với bài toán nhận diện tiếng nói từng từ, đầu ra sẽ là các từ riêng lẻ. Đối với bài toán nhận diện tiếng nói thành câu hoặc đoạn văn, đầu ra sẽ là các câu hoặc đoạn văn tương ứng.
- Định danh người nói: Đây là thông tin đầu ra chính của bài toán nhận diện người nói. Khi hệ thống nhận diện được người nói, nó sẽ trả về một định danh tương ứng với người đó. Định danh này có thể được đại diện bằng tên người nói hoặc một mã số duy nhất. 
- Các thông tin khác về người nói: Ngoài định danh, các thông tin khác về người nói, chẳng hạn như giới tính, độ tuổi, ngôn ngữ, vùng miền, và những đặc điểm khác của giọng nói cũng có thể được trả về làm kết quả của bài toán nhận diện người nói.

*Đánh giá*: 

Để đánh giá bài toán nhận diện người nói Speaker Recognition bằng độ đo Accuracy (Độ chính xác) là một độ đo thông thường được sử dụng để đánh giá hiệu suất của các hệ thống xử lý tiếng nói. Độ chính xác đo lường tỷ lệ giữa số lần phân loại chính xác và tổng số lần phân loại.

# Accuracy = number of right prediction / number of data
<!-- ![](https://i.imgur.com/qkrbm8L.png) -->

Với
- Number of right prediction: là số dự đoán đúng.
- Number of data: là số mẫu của dữ liệu.

*1.2.2.2 Nhận dạng giọng nói (Speech to text)*: nhận diện tiếng nói là một bài toán trong lĩnh vực xử lý tiếng nói, trong đó mục tiêu là phân loại âm thanh đầu vào thành các từ hoặc câu nói tương ứng với ngôn ngữ đã cho trước. Bài toán này có rất nhiều ứng dụng trong cuộc sống thực, chẳng hạn như hệ thống nhận diện giọng nói để mở khóa điện thoại, hệ thống điều khiển bằng giọng nói, hoặc hệ thống chuyển đổi giọng nói thành văn bản. 

*Dữ liệu đầu vào* của bài toán nhận diện tiếng nói là tín hiệu âm thanh thu được từ microphone hoặc từ các tệp âm thanh, có thể là các file WAV. Tuy nhiên, trước khi sử dụng cho bài toán nhận diện tiếng nói, tín hiệu âm thanh thường phải trải qua một quá trình tiền xử lý để loại bỏ nhiễu và chuẩn hóa.

*Dữ liệu đầu ra*: nhận diện tiếng nói là các chuỗi ký tự tương ứng với những gì được nói trong tín hiệu âm thanh đầu vào. Đối với bài toán nhận diện tiếng nói từng từ, đầu ra sẽ là các từ riêng lẻ. Đối với bài toán nhận diện tiếng nói thành câu hoặc đoạn văn, đầu ra sẽ là các câu hoặc đoạn văn tương ứng

*Đánh giá*:
**WER (Word Error Rate)** là một độ đo được sử dụng để đánh giá hiệu suất của các hệ thống nhận diện tiếng nói. WER được tính bằng tỉ lệ giữa số lượng từ bị sai so với tổng số từ trong câu hoặc tài liệu. 

# WER = ( S + D + I) / N

<!-- ![](https://i.imgur.com/Ds2VpUm.png) -->

Trong đó: 
- S là số lượng từ được nhận diện đúng và đúng vị trí 
- D là số lượng từ trong chuỗi đánh nhãn mà hệ thống không nhận diện được 
- I là số lượng từ trong chuỗi kết quả mà không có trong chuỗi đánh nhãn 
- N là tổng số từ trong chuỗi đánh nhãn

**BLEU (BiLingual Evaluation Understudy)** là một độ đo được sử dụng để đánh giá chất lượng của các bản dịch máy. Nó dựa trên việc so sánh giữa các câu dịch được tạo ra bởi hệ thống với các câu dịch chuẩn được xác định trước đó bởi con người.
BLEU được tính bằng cách so sánh độ chính xác của các từ trong các câu dịch. Để tính toán BLEU, cần có một bộ dữ liệu kiểm tra bao gồm các cặp câu văn bản trong ngôn ngữ nguồn và ngôn ngữ đích, cùng với các câu dịch chuẩn được đánh giá bởi con người. Sau đó, hệ thống dịch máy được sử dụng để tạo ra các câu dịch mới cho các cặp câu trong bộ dữ liệu kiểm tra. Các câu dịch này sau đó được so sánh với các câu dịch chuẩn bằng BLEU.

# BLUE = BP x exp⁡( ∑ ( i - N ) Wi ×log⁡(Pn))

<!-- ![](https://i.imgur.com/9RzyEI1.png) -->

**BP**: là chỉ số brevity penalty, bao gồm các tham số: c (candidate translation) là tổng số lượng các từ trong bản dịch cần đánh giá từ trong hệ thống dịch máy, r (reference translation) là tổng số lượng các từ trong bản dịch tham khảo của con người. BP được tính bằng công thức sau:

# BP= 1 if (c > r) else (c ≤ r)BP = e^((1-r/c))

<!-- ![](https://i.imgur.com/ckLrMBn.png) -->

**Pn**: là chỉ số modified n-gram precision, biểu diễn mức độ trùng khớp giữa văn bản cần đánh giá của hệ thống dịch máy so với văn bản tham khảo từ con người. được tính như sau:
# Pn = m / Wt
<!-- ![](https://i.imgur.com/6vOYvg0.png) -->

*1.2.3 Một số thách thức và khó khăn trong quá trình nghiên cứu và phát triển hệ thống nhận dạng tiếng nói*

Trong quá trình nhận dạng giọng nói, nhiều thách thức phải được giải quyết. Các thách thức này bao gồm: độ chính xác thấp do ảnh hưởng của tiếng ồn và các biến thể của giọng nói, sự khác biệt về ngữ điệu và phong cách nói của mỗi người, và khả năng tương tác giữa người dùng và hệ thống. Do đó, để tăng độ chính xác và độ tin cậy của hệ thống nhận dạng giọng nói, cần phải sử dụng các kỹ thuật tiên tiến trong xử lý tín hiệu số và trích xuất đặc trưng giọng nói, cùng với các thuật toán máy học để phân loại và nhận dạng giọng nói.

- *Ảnh hưởng của tiếng ồn:* Tiếng ồn là một trong những yếu tố gây ảnh hưởng lớn đến chất lượng của giọng nói. Trong một môi trường ồn ào, tín hiệu giọng nói sẽ bị méo mó và làm giảm độ chính xác của hệ thống nhận dạng.
- *Sự khác biệt về ngữ điệu và phong cách nói của mỗi người:* Mỗi người có một cách nói riêng biệt, với sự khác biệt về cách phát âm, tốc độ, âm lượng, v.v. Khác biệt này có thể làm cho hệ thống nhận dạng khó khăn hơn khi phải xử lý nhiều biến thể của giọng nói.
- *Thiếu dữ liệu:* Việc thu thập và xây dựng tập dữ liệu giọng nói là một công việc phức tạp và tốn kém. Nếu tập dữ liệu quá nhỏ hoặc không đủ đại diện, hệ thống nhận dạng có thể không đạt được độ chính xác mong muốn.
- *Khả năng tương tác giữa người dùng và hệ thống:* Hệ thống nhận dạng giọng nói thường được sử dụng trong các ứng dụng tương tác người-máy. Tuy nhiên, khả năng tương tác giữa người dùng và hệ thống có thể bị gián đoạn bởi các yếu tố như thời gian trễ, sai sót nhận dạng, hoặc các lỗi khác trong quá trình truyền dữ liệu.
- *Độ chính xác của giọng nói:* Giọng nói của người già thường có nhiều biến thể, có độ chính xác thấp hơn so với giọng nói của người trẻ tuổi. Điều này làm tăng độ khó trong việc phân loại và nhận dạng giọng nói.

Một số mục tiêu nghiên cứu cụ thể mà bài báo cáo sẽ tập trung vào: cải thiện độ chính xác của hệ thống nhận dạng giọng nói, tăng khả năng xử lý các trường hợp giọng nói đa dạng, giảm ảnh hưởng của tiếng ồn, và tăng tính linh hoạt của hệ thống trong việc định danh người nói và phân tích nội dung câu nói.

*2.2.1 Tổng quan về Transformer:*

Transformer ra đời kế thừa ý tưởng từ self attention từ LSTMN, loại bỏ hoàn toàn tính tuần tự phụ thuộc hoàn toàn vào cơ chế attention để tính toán ra được mối tương quan giữa input và output.
- Self-attention, đôi khi được gọi là intra-attention, là một cơ chế attention liên quan đến các vị trí khác nhau của một chuỗi để tính toán biểu diễn của chuỗi.
- Multi-head attention là một mô-đun cho các cơ chế attention chạy song song qua một cơ chế attention nhiều lần. Các đầu ra attention độc lập sau đó được nối và chuyển đổi tuyến tính thành chiểu như mong muốn.

Ở mỗi bước decoder, ta tính attention như sau:
- Nhận đầu vào attention: trạng thái bộ giải mã ht và tất cả các trạng thái encoder s1, s2, s3, s4, …, sm.
- Tính attention score: ta áp dụng một chức năng Attention nhận một trạng thái bộ giải mã và một trạng thái bộ mã hóa và trả về một giá trị vô hướng score(ht, sk).
- Tính attention weight: phân phối xác suất - softmax được áp dụng cho attention score.
- Tính attention output: tổng trọng số của các trạng thái bộ mã hóa với attention weight.

*2.2.2 Kiến trúc Transformer*

Transformer dựa theo kiến trúc Encoder-Decoder tránh lặp lại và thay vào đó hoàn toàn dựa vào một cơ chế Attention để thu hút sự phụ thuộc toàn cầu giữa đầu vào và đầu ra.
Transformer cho phép song song hóa nhiều hơn đáng kể và có thể đạt đến trình độ mới về chất lượng bản dịch sau khi được đào tạo trong ít nhất mười hai giờ trên tám GPU P100.

![](https://i.imgur.com/KXdcYEB.png)

Mô hình sẽ bao gồm 2 phase.
- Encoder: Bao gồm 6 layers liên tiếp nhau. Mỗi một layer sẽ bao gồm một sub-layer là Multi-Head Attention kết hợp với fully-connected layer như mô tả ở nhánh encoder bên trái của hình vẽ. Kết thúc quá trình encoder ta thu được một vector embedding output cho mỗi từ.
- Decoder: Kiến trúc cũng bao gồm các layers liên tiếp nhau. Mỗi một layer của Decoder cũng có các sub-layers gần tương tự như layer của Encoder nhưng bổ sung thêm sub-layer đầu tiên là Masked Multi-Head Attention có tác dụng loại bỏ các từ trong tương lai khỏi quá trình attention.
- Encoder-decoder attention: Mỗi block chứa Multihead-attention, Add & Norm, Feed Forward, Add & Norm. Attention được sử dụng ở đây không phải dạng self attention do queries nhận output của lớp decoder phía trước trong khi đó keys và values nhận cùng giá trị từ output của phần decoder.Attention ở đây tương tự như việc sử dụng attention trong mô hình encoder-decoder truyền thống vậy.

Sở dĩ được gọi là encoder-decoder attention vì đây là kiến trúc attention tương tác giữa các véc tơ embedding của encoder và decoder. véc tơ context được tính toán trên encoder đã được tính tương quan với véc tơ decoder nên sẽ có ý nghĩa giải thích bối cảnh của từ tại vị trí time step decoder tương ứng. Sau khi kết hợp giữa vectơ context và véc tơ decoder ta sẽ project tiếp qua một fully connected layer để tính phân phối xác suất cho output

*2.2.3 Tổng quan về CNN:*

Convolutional Neural Network (CNN hoặc ConvNet) được tạm dịch là: Mạng nơ ron tích chập. Đây được xem là một trong những mô hình của Deep Learning – tập hợp các thuật toán để có mô hình dữ liệu trừu tượng hóa ở mức cao bằng cách sử dụng nhiều lớp xử lý cấu trúc phức tạp. Hiểu đơn giản, CNN là một lớp của mạng nơ-ron sâu , được áp dụng phổ biến nhất để phân tích hình ảnh trực quan.

![](https://i.imgur.com/YtiAwV0.png)

CNN (Convolutional Neural Network) là một loại kiến trúc mạng thần kinh sử dụng trong việc xử lý tiếng nói. Trong CNN, các lớp chính được sử dụng bao gồm:

![](https://i.imgur.com/CHDKpTN.png)

- **Convolutional layer:** lớp tích chập là lớp cơ bản của CNN, nó áp dụng một bộ lọc (kernel) để tính toán tích chập giữa đầu vào và bộ lọc. Đầu ra của lớp tích chập được gọi là feature map, nó bao gồm các đặc trưng cục bộ của ảnh đầu vào. 
- **Pooling layer:** lớp gộp được sử dụng để giảm kích thước của feature map bằng cách lấy giá trị trung bình hoặc giá trị lớn nhất từ một vùng của feature map. Lớp gộp giúp giảm thiểu số lượng tham số của mô hình và giảm overfitting. 
- **Activation layer:** lớp kích hoạt áp dụng một hàm phi tuyến (ví dụ như ReLU) lên feature map để tạo ra feature map mới với các giá trị dương. 
- **Fully Connected layer:** lớp kết nối đầy đủ được sử dụng để kết nối các feature map đến một tập hợp đầy đủ các đơn vị ẩn, sau đó đưa ra đầu ra dưới dạng xác suất cho các lớp phân loại.
- **Convolution layer:** đây là lớp quan trọng nhất của CNN, lớp này có nhiệm vụ thực hiện mọi tính toán. Những yếu tố quan trọng của một convolutional layer là: filter map, stride, padding, feature map.
> - CNN: sử dụng các filter để áp dụng vào vùng của hình ảnh. Những filter map này được gọi là ma trận 3 chiều, mà bên trong nó là các con số và chúng là parameter.
> - Stride: có nghĩa là khi bạn dịch chuyển filter map theo pixel dựa vào giá trị từ trái sang phải. Và sự chuyển dịch này chính là Stride. 
> - Padding: Là các giá trị 0 được thêm vào với lớp input. 
> - Feature map: Nó thể hiện kết quả của mỗi lần filter map quét qua input. Sau mỗi lần quét sẽ xảy ra quá trình tính toán.

- **Pooling layer:** thường được dùng giữa các convolutional layer, để giảm kích thước dữ liệu nhưng vẫn giữ được các thuộc tính quan trọng. Việc giảm kích thước dữ liệu giúp giảm các phép tính toán trong model.
- **Relu layer:** là một hàm kích hoạt trong Neural Network. Chúng ta biết đến hàm kích hoạt này với một tên gọi khác là Activation Function. Nhiệm vụ chính của hàm kích hoạt là mô phỏng lại các Neuron có tỷ lệ truyền xung qua Axon. Trong đó, hàm kích hoạt sẽ bao gồm các hàm cơ bản như: Sigmoid, Tanh, Relu, Leaky Relu, Maxout.
- **Fully Connected layer:** là một lớp trong kiến trúc mạng nơ-ron nhân tạo (artificial neural network), nơi mỗi đơn vị (neuron) trong lớp này được kết nối đầy đủ với tất cả các đơn vị của lớp trước và sau nó. Lớp này còn được gọi là lớp kết nối đầy đủ (dense layer). Fully connected layer thường được kết hợp với các lớp khác như lớp tích chập (convolutional layer) và lớp gộp (pooling layer) để tạo thành kiến trúc mạng CNN (Convolutional Neural Network).

```python
# biểu diễn câu chữ dưới dạng vector
class TokenEmbedding(layers.Layer):
    def __init__(self, num_vocab=1000, maxlen=100, num_hid=64):
        # khởi tạo một ma trận số nguyên với phần tử (word index) lớn nhất nhỏ hơn num_vocab
        super().__init__()
        self.emb = tf.keras.layers.Embedding(num_vocab, num_hid)
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=num_hid)

    def call(self, x):
        maxlen = tf.shape(x)[-1]
        x = self.emb(x)
        # đánh dấu vị trí của từ trong câu, vì một từ ở vị trí khác nhau của câu có thể sẽ mang ý nghĩa khác nhau
        positions = tf.range(start=0, limit=maxlen, delta=1)
        positions = self.pos_emb(positions)
        return x + positions


class SpeechFeatureEmbedding(layers.Layer):
    def __init__(self, num_hid=64, maxlen=100):
        super().__init__()
        # Từ vector input, tạo ra một convolution kernel được xoay vòng để tạo ra nhiều đầu ra
        self.conv1 = tf.keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )
        self.conv2 = tf.keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )
        self.conv3 = tf.keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=num_hid)

    def call(self, x):
        """
            input: một mảng x
            output: là một mảng đã được downsample 3 lần, mỗi lớp ta sẽ chia dữ liệu ra còn một nữa
        """
        x = self.conv1(x)
        # x = self.conv2(x)
        return self.conv3(x)
```

```python
# Định nghĩa lớp Encode của model Transformer
class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads, feed_forward_dim, rate=0.1):
        # sử dụng multi-head attention để tìm sự liên kết giữa các từ trong câu
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = keras.Sequential(
            [
                layers.Dense(feed_forward_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )
        # Thông tin ban đầu trước khi được self-attention (vector x) bổ sung thêm với 
        # thông tin đã được self-attention (vector z) - là mối quan hệ giữa các từ trong câu
        # sau đó mô hình thực hiện Layer Normalization
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        # Giai đoạn này được gọi là Residuals
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        # chuẩn hóa inputs
        out1 = self.layernorm1(inputs + attn_output)
        # Sau khi được Normalize, các vectors z được đưa qua mạng fully connected 
        # trước khi đẩy qua Decoder. Vì các vectors này không phụ thuộc vào nhau 
        # nên ta có thể tận dụng được tính toán song song cho cả câu.
        ffn_output = self.ffn(out1)  # feed-forward network
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)
```

```python
# Quá trình decode của model Transformer về cơ bản là giống với encode:
class TransformerDecoder(layers.Layer):
    def __init__(self, embed_dim, num_heads, feed_forward_dim, dropout_rate=0.1):
        super().__init__()
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm3 = layers.LayerNormalization(epsilon=1e-6)
        self.self_att = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim
        )
        self.enc_att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.self_dropout = layers.Dropout(0.5)
        self.enc_dropout = layers.Dropout(0.1)
        self.ffn_dropout = layers.Dropout(0.1)
        self.ffn = keras.Sequential(
            [
                layers.Dense(feed_forward_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )

    def causal_attention_mask(self, batch_size, n_dest, n_src, dtype):
        # nếu ta sử dụng Multi-head attention trên cả câu như ở Encoder, 
        # Decoder sẽ nhận luôn từ tiếp theo mà nó cần dịch. Để ngăn điều đó, 
        # khi Decoder dịch đến từ thứ i, phần sau của câu sẽ bị che lại (masked) 
        # và Decoder chỉ được phép nhận thấy phần nó đã dịch trước đó.
        i = tf.range(n_dest)[:, None]
        j = tf.range(n_src)
        m = i >= j - n_src + n_dest
        mask = tf.cast(m, dtype)
        mask = tf.reshape(mask, [1, n_dest, n_src])
        mult = tf.concat(
            [tf.expand_dims(batch_size, -1), tf.constant([1, 1], dtype=tf.int32)], 0
        )
        # trả về mảng mới với các chiều dữ liệu mới dựa theo mult
        return tf.tile(mask, mult)

    def call(self, enc_out, target): # thiết kế lớp cho decoder
        input_shape = tf.shape(target)
        batch_size = input_shape[0]
        seq_len = input_shape[1]
        causal_mask = self.causal_attention_mask(batch_size, seq_len, seq_len, tf.bool)
        target_att = self.self_att(target, target, attention_mask=causal_mask)
        target_norm = self.layernorm1(target + self.self_dropout(target_att))
        enc_out = self.enc_att(target_norm, enc_out)
        enc_out_norm = self.layernorm2(self.enc_dropout(enc_out) + target_norm)
        # feed-forward network
        ffn_out = self.ffn(enc_out_norm)
        ffn_out_norm = self.layernorm3(enc_out_norm + self.ffn_dropout(ffn_out))
        return ffn_out_norm
```

## Hoàn thành model Transformer

Model của nhóm em lấy phổ của âm thanh làm đầu vào và dự đoán thành một chuỗi ký tự

Trong quá trình huấn luyện, nhóm em cung cấp cho decoder một chuỗi các target character ở bên trái như input

Trong suốt quá trình suy luận, decoder sử dụng quá khứ của chính nó để dự đoán cho các token tiếp theo

```python
class Transformer(keras.Model):
    # truyền vào các tham số, các nút trong các tầng ẩn
    def __init__(
        self,
        num_hid=64,
        num_head=2, # chiều của các embedding
        num_feed_forward=128,
        source_maxlen=100,
        target_maxlen=100,
        num_layers_enc=4,
        num_layers_dec=1,
        num_classes=10, # class
    ):
        super().__init__()
        self.loss_metric = keras.metrics.Mean(name="loss")
        self.num_layers_enc = num_layers_enc
        self.num_layers_dec = num_layers_dec
        self.target_maxlen = target_maxlen
        self.num_classes = num_classes

        self.enc_input = SpeechFeatureEmbedding(num_hid=num_hid, maxlen=source_maxlen)
        self.dec_input = TokenEmbedding(
            num_vocab=num_classes, maxlen=target_maxlen, num_hid=num_hid
        )

        self.encoder = keras.Sequential(
            [self.enc_input]
            + [
                TransformerEncoder(num_hid, num_head, num_feed_forward)
                for _ in range(num_layers_enc)
            ]
        )

        for i in range(num_layers_dec):
            setattr(
                self,
                f"dec_layer_{i}",
                TransformerDecoder(num_hid, num_head, num_feed_forward),
            )

        self.classifier = layers.Dense(num_classes)

    def decode(self, enc_out, target):
        # Dự đoán từng từ trong câu
        y = self.dec_input(target)
        for i in range(self.num_layers_dec):
            y = getattr(self, f"dec_layer_{i}")(enc_out, y)
        return y

    def call(self, inputs):
        """
            input: đưa vào các ký tự
            output: nhãn của ký tự đó
        """
        source = inputs[0]
        target = inputs[1]
        x = self.encoder(source)
        y = self.decode(x, target)
        return self.classifier(y)

    @property
    def metrics(self):
        return [self.loss_metric]

    def train_step(self, batch):
        # Xử lý một batch trong model.fit()
        source = batch["source"]
        target = batch["target"]
        dec_input = target[:, :-1]
        dec_target = target[:, 1:]
        with tf.GradientTape() as tape:
            preds = self([source, dec_input])
            one_hot = tf.one_hot(dec_target, depth=self.num_classes)
            mask = tf.math.logical_not(tf.math.equal(dec_target, 0))
            loss = self.compiled_loss(one_hot, preds, sample_weight=mask)
        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))
        self.loss_metric.update_state(loss)
        return {"loss": self.loss_metric.result()}

    def test_step(self, batch):
        """
            input: truyền vào batch của data
            output: trả về dict keylaf loss
        """
        source = batch["source"]
        target = batch["target"]
        dec_input = target[:, :-1]
        dec_target = target[:, 1:]
        preds = self([source, dec_input])
        one_hot = tf.one_hot(dec_target, depth=self.num_classes)
        mask = tf.math.logical_not(tf.math.equal(dec_target, 0))
        loss = self.compiled_loss(one_hot, preds, sample_weight=mask)
        self.loss_metric.update_state(loss)
        return {"loss": self.loss_metric.result()}

    def generate(self, source, target_start_token_idx):
        # Thực hiện suy luận trên một loạt đầu vào (batch of input) bằng cách sử dụng giải thuật greedy
        bs = tf.shape(source)[0]
        enc = self.encoder(source)
        dec_input = tf.ones((bs, 1), dtype=tf.int32) * target_start_token_idx
        dec_logits = []
        for i in range(self.target_maxlen - 1):
            dec_out = self.decode(enc, dec_input)
            logits = self.classifier(dec_out)
            logits = tf.argmax(logits, axis=-1, output_type=tf.int32)
            last_logit = tf.expand_dims(logits[:, -1], axis=-1)
            dec_logits.append(last_logit)
            dec_input = tf.concat([dec_input, last_logit], axis=-1)
        return dec_input
```

```python
# VectorizeChar để chuyển các char trong text thành vector id ký tự
class VectorizeChar:
    def __init__(self, max_len=50):
        self.vocab = (
            ["-", "#", "<", ">"]
            + [chr(i + 96) for i in range(1, 27)]
            + [" ", ".", ",", "?", "â", "ă", "ê", "ô", "ơ", "ư", "à", "ằ", "ầ", "è", "ề", "ò", "ồ", "ờ", "ù", "ừ", "ì", "ỳ", "á", "ắ", "ấ", "é", "ế", "ó", "ố", "ớ", "ú", "ứ", "í", "ý", "ả", "ẳ", "ẩ", "ẻ", "ể", "ỏ", "ổ", "ở", "ủ", "ử", "ỉ", "ỷ", "ã", "ẵ", "ẫ", "ẽ", "ễ", "õ", "ỗ", "ỡ", "ũ", "ữ", "ĩ", "ỹ", "ạ", "ặ", "ậ", "ẹ", "ệ", "ọ", "ộ", "ợ", "ụ", "ự", "ị", "ỵ","đ"]
        )
        self.max_len = max_len
        # dữ liệu trong char_to_idx là: vocab: id
        self.char_to_idx = {}
        for i, ch in enumerate(self.vocab):
            self.char_to_idx[ch] = i

    def __call__(self, text):
        """
            input: text
            output: danh sách id của các ký tự (char)
        """
        # viết thường hết đoạn text truyền vào
        text = text.lower()
        # lấy kích thước các ký tự trong text không quá maxlen-2
        text = text[: self.max_len - 2]
        # đưa text vào trong ngoặc nhọn
        text = "<" + text + ">"
        # dùng để padding đoạn text
        pad_len = self.max_len - len(text)
        # nếu ký tự không có trong list thì mặc định là 1, sau khi bỏ đi phần sau của text thì chúng ta sẽ thêm zero array để padding
        return [self.char_to_idx.get(ch, 1) for ch in text] + [0] * pad_len

    def get_vocabulary(self):
        # output: danh sách vocab
        return self.vocab
```

```python
# Chuyển data về vector của text trong data dưới dạng TensorSliceDataset
def create_text_ds(data):
    """
        input: data là một danh sách các dict có cấu trúc: 
                audio: đường dẫn đến file audio wav,
                text: chứa nội dung text (không quá maxlen)
        output: các vector của text trong data dưới dạng TensorSliceDataset 
    """
    # các text trong data
    texts = [_["text"] for _ in data]
    # chuyển thành vector các texts
    text_ds = [vectorizer(t) for t in texts]
    text_ds = tf.data.Dataset.from_tensor_slices(text_ds)
    return text_ds

import soundfile as sf
def path_to_audio(path):
    """
        input: đường dẫn của file audio wav
        output: x là phổ của audio đã được chuẩn hóa có shape = (pad_len, ?)
    """
    # đọc tệp WAV bằng thư viện soundfile
    audio, sr = sf.read(path)
    # chuẩn hóa giá trị âm thanh
    audio = audio / 128.0

    # chuyển shape của audio sang 1d từ (?,) thành (1, ?)
    stfts = tf.signal.stft(audio, frame_length=200, frame_step=80, fft_length=256)
    # lấy căn trị tuyệt đối của phổ
    x = tf.math.pow(tf.abs(stfts), 0.5)
    # chuẩn hóa theo trung bình và lộ lệch chuẩn của từng mảng phần tử trong x
    means = tf.math.reduce_mean(x, 1, keepdims=True)
    stddevs = tf.math.reduce_std(x, 1, keepdims=True) + 0.000001
    x = (x - means) / stddevs
    audio_len = tf.shape(x)[0]
    #Giá trị phổ âm thanh được thêm padding ở cuối mảng để đạt được độ dài bằng 550 (pad_len), tương ứng với khoảng thời gian 2 giây
    pad_len = 550
    paddings = tf.constant([[0, pad_len], [0, 0]])
    x = tf.pad(x, paddings, "CONSTANT")[:pad_len, :]
    return x 

#Chuyển audio sang phổ của audio đó và đã được chuẩn hóa
def create_audio_ds(data):
    """
        input: data là một danh sách các dict có cấu trúc: 
                audio: đường dẫn đến file audio wav,
                text: chứa nội dung text (không quá maxlen)
        output: các phổ của audio đã được chuẩn hóa
    """
    flist = [_["audio"] for _ in data]

    audio_ds = [path_to_audio(i) for i in flist]
    audio_ds = tf.data.Dataset.from_tensor_slices(audio_ds)
    return audio_ds

#Tạo Tập dữ liệu bao gồm phổ và vector của Text
def create_tf_dataset(data, bs=4):
    """
        input: 
            data: một danh sách các dict có cấu trúc: 
                audio: đường dẫn đến file audio wav,
                text: chứa nội dung text (không quá maxlen)
                bs: là batch size
        output: là một PrefetchDataset như dict có cấu trúc: 
            source: chứa tất cả phổ của âm thanh đã được chuẩn hóa
            target: chứa tất cả vector của text
    """
    audio_ds = create_audio_ds(data)
    text_ds = create_text_ds(data)

    # đùng để loại bỏ đi những mảng có nan bên trong
    ds = tf.data.Dataset.zip((audio_ds, text_ds))#.filter(filter_nan)
    # chuyển thành dict
    ds = ds.map(lambda x, y: {"source": x, "target": y})
    ds = ds.shuffle(buffer_size=len(ds), seed=42, reshuffle_each_iteration=False)
    ds = ds.batch(bs)
    ds = ds.prefetch(tf.data.AUTOTUNE)
    return ds
```

## Định nghĩa một callbacks để hiển thị các dự đoán

```python
class DisplayOutputs(keras.callbacks.Callback):
    def __init__(
        self, batch, idx_to_token, target_start_token_idx=27, target_end_token_idx=28
    ):
        # Hiển thị một batch kết quả đầu ra sau mỗi epoch

        # Các tham số:
        #     batch: Một batch thử nghiệm chứa các khóa "source" và "target"
        #     idx_to_token: Danh sách chứa các mã thông báo từ vựng tương ứng với các chỉ số của chúng
        #     target_start_token_idx: Chỉ mục mã thông báo bắt đầu trong từ vựng target
        #     target_end_token_idx: Chỉ mục mã thông báo kết thúc trong từ vựng target
        self.batch = batch
        self.target_start_token_idx = target_start_token_idx
        self.target_end_token_idx = target_end_token_idx
        self.idx_to_char = idx_to_token

    def on_epoch_end(self, epoch, logs=None):
        # trong mỗi epoch nếu %5 != 0 thì không làm gì cả
        # hàm này sẽ in ra quá kết quả dự đoán và kết quả thực tế của model trong quá trình training
        if epoch % 5 != 0:
            return
        source = self.batch["source"]
        target = self.batch["target"].numpy()
        bs = tf.shape(source)[0]
        preds = self.model.generate(source, self.target_start_token_idx)
        preds = preds.numpy()
        for i in range(bs):
            target_text = "".join([self.idx_to_char[_] for _ in target[i, :]])
            prediction = ""
            for idx in preds[i, :]:
                prediction += self.idx_to_char[idx]
                if idx == self.target_end_token_idx:
                    break
            print(f"target:     {target_text.replace('-','')}")
            print(f"prediction: {prediction}\n")
```

```python
import nltk
nltk.download('punkt')
import warnings
warnings.filterwarnings("ignore")
from nltk.translate.bleu_score import sentence_bleu
from nltk.tokenize import word_tokenize
!pip install jiwer --quiet
from jiwer import wer

def bleu_score(pred, target):
    return sentence_bleu([word_tokenize(target)], word_tokenize(pred), weights=(1, 0, 0, 0))
```
